```python
cities = ["Chicago", "Springfield","Cincinnati","Cleveland", "San+francisco", "Los+angeles", "Birmingham", "Dothan", "New+york", 
          "Syracuse", "Tucson", "Flagstaff", "Seattle", "Kennewick", "Boston", "Pittsfield", "Jacksonville", "Miami"]
start = "2013-01-01"
end = "2017-12-31"
City = []
Dates = []
MaxTemps = []
```

```python
for city in cities:
    for i in range(29):
        url = "http://api.worldweatheronline.com/premium/v1/past-weather.ashx?"
        query_url =f"{url}key=8e7a93ba07e84200a8a180607180508&q={city}&format=json&date={start}&enddate={end}&tp=24"
        response = requests.get(query_url).json()
        City.append(response["data"]["request"][i]["query"])
        Dates.append(response["data"]["weather"][i]["date"])
        MaxTemps.append(response["data"]["weather"][i]["maxtempF"])
        
```

```python
Weatherdict = {"City": City,
              "Date": Dates,
              "Maximum Temperature": MaxTemps}

Weather_data = pd.DataFrame(Weatherdict)

print(Weather_data.head())

Weather_data.to_csv("Output/weatherdata2013.csv")
```
